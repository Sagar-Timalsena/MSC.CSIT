```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
bias = 0
alpha = 1
def perceptron_training(x, w, t):
  global bias
  for i in range(0, len(x)):
    v = sum(x[i]*w) + bias
    y = hard_limiter(v)
    weight_update = alpha * (t[i] - y) * x[i]
    w = np.add(w, weight_update)
    bias_update = alpha * (t[i] - y)
    bias = bias + bias_update
    return w
def predict_perception(x, w):
  z = x * w
  v = sum(z) - bias
  y = hard_limiter(v)
  return y
def hard_limiter(e):
  if(e > 0):
    return 1
  elif (e < 0):
    return -1
  else:
    return 0




training_input = np.array([[5.9, 75],[5.8, 86],[5.2, 50],[5.4, 55],[6.1, 85],[5.5, 62]])
training_output = np.array([1, 1, -1, -1, 1, -1])
# test_input = np.array([[6, 82]])
test_input = np.array([[5.3, 52]])
scaler = MinMaxScaler()
transform_training_input = scaler.fit_transform(training_input)
transform_test_input = scaler.transform(test_input)
wt = np.array([0, 0])
print("Training the perceptron...")
for i in range(1, 5):
  print("Epoch: ", i)
  wt = perceptron_training(transform_training_input, wt, training_output)
  print("Weight after epoch ", i)
  print(wt)
print("Bias: ", bias)
print("Test data: ", transform_test_input)
transform_test_input = transform_test_input.flatten()
y = predict_perception(transform_test_input, wt)
if(y == 1):
  print("Predicted class is : male")
else:
  print("Predicted class is : female")
```
