```python
import numpy as np
def MinMaxScalar(data):
  maxValue = np.amax(data)
  minValue = np.amin(data)
  print("Data after scale is: ")
  return ((np.array(data) - minValue) / (maxValue - minValue))

data = [[10 , 10],[5 , 5 ],[10 , 13 ],[15, 20]]
scaledData = MinMaxScalar(data)
print(scaledData)
```
